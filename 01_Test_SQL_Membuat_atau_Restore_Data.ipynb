```python
# import sqlite
import sqlite3
```

```python
# Membuat database bernama example, dan membuat connection bernama connection1
connection1=sqlite3.connect("example.db")
```

```python
# Membuat dataframe dengan dummy data
import pandas as pd
df=pd.DataFrame({'col': [1,2,3]})
```

```python
df
```

```python
# Memasukkan data frame, dengan nama test_table ke dalam database kita dengan menggunakan koneksi data connection 1
df.to_sql('test_table', connection1)
```

```python
# coba query data dari table test_tabel di database kita
pd.read_sql_query("SELECT * from test_table", connection1)
```
